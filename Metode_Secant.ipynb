```python
 import numpy as np
 import matplotlib.pyplot as plt

 # Fungsi yang ingin dicari akarnya
 def f(x):
    return 20*x**4 - 2*x**3 + 2*x - 35  # Fungsi contoh: f(x)

 # Metode Secant
 def secant_method(f, x0, x1, tol=0.01, max_iter=100):
     iterasi = 0
     x_values = [x0, x1]  # Menyimpan nilai x untuk grafik dan iterasi
     print(f"Iterasi 0: x0 = {x0}, f(x0) = {f(x0)}")
     print(f"Iterasi 1: x1 = {x1}, f(x1) = {f(x1)}")
     while iterasi < max_iter:
         # Hitung nilai f(x) di kedua titik
         f_x0 = f(x0)
         f_x1 = f(x1)
         # Formula metode secant
         x2 = x1 - f_x1 * (x1 - x0) / (f_x1 - f_x0)
         # Menambahkan nilai x2 ke daftar untuk grafik
         x_values.append(x2)
         # Menampilkan iterasi
         print(f"Iterasi {iterasi + 2}: x2 = {x2}, f(x2) = {f(x2)}")
         # Jika perubahan antara x1 dan x2 cukup kecil, berhenti
         if abs(x2 - x1) < tol:
             break
         # Update x0 dan x1 untuk iterasi berikutnya
         x0, x1 = x1, x2
         iterasi += 1

     return x2, iterasi, x_values

 # Parameter awal
 x0 = 2.0  # Titik awal x0
 x1 = 3.0  # Titik awal x1

 # Menjalankan metode secant
 akar, iterasi, x_values = secant_method(f, x0, x1)

 # Menampilkan hasil akar dan jumlah iterasi
 print(f"\nAkar yang ditemukan: {akar}")
 print(f"Jumlah iterasi: {iterasi}")

 # Membuat grafik fungsi dan titik iterasi
 x = np.linspace(0, 5, 400)
 y = f(x)

 # Plot grafik fungsi
 plt.figure(figsize=(8, 6))
 plt.plot(x, y, label="f(x) = x^3 - 6x^2 + 11x - 6", color="blue")
 plt.axhline(0, color='black',linewidth=1)
 plt.axvline(0, color='black',linewidth=1)
 # Plot titik-titik iterasi metode secant
 for i, xi in enumerate(x_values):
     plt.plot(xi, f(xi), 'ro')  # Titik merah untuk iterasi
     plt.text(xi, f(xi), f' Iter {i}', color='red')
 # Menambahkan label dan title
 plt.title('Grafik Fungsi dan Iterasi Metode Secant')
 plt.xlabel('x')
 plt.ylabel('f(x)')
 plt.legend()
 plt.grid(True)
 plt.show()
```
